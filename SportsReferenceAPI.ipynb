```python
#import NBA package
from sportsreference.nba.teams import Teams
```

```python
#print NBA teams and their respective wins + loses
teams = Teams()
for team in teams:
    print (team.name)
```

```python
#Import NHL package
from sportsreference.nhl.teams import Teams
```

```python
teams = Teams()
for team in teams:
    print(team.name, team.wins, team.losses)  # Prints the team's name, their current 2020 win-loss record
```

```python
#printing win-loss from the 2019 hockey season
teams = Teams('2019')
for team in teams:
    print(team.name, team.wins, team.losses)
```

```python
#print teams with Pandas dataframe integration
teams = Teams()
for team in teams:
    print(team.name)
    print(team.dataframe)
```
